# Data Visualization

The following notebook will present some visualization to understand the characteristics of certain important Datasets.

First the the basic Dataset **430650** will be visualized. This dataset represents the default data within this thesis.

```python
# Define Path to the Data Repository
data_path = 'C:/Users/lanza/Master_Thesis_EL/Integrated-vs-Seperated-Master-Thesis/04 Data/'
```

```python
from IvsS_Evaluation_Utils import load_generated_data

X_train, y_train, X_val, y_val, X_test, y_test = load_generated_data(path=data_path+'Size/Low_variance/set_430650_data.h5', multi=True)
```

```python
import pandas as pd

# Convert the data into pandas DataFrames for easier manipulation
df_train = pd.DataFrame(X_train)
for i in range(y_train.shape[1]):
    df_train[f'target_{i+1}'] = y_train[:, i]

df_val = pd.DataFrame(X_val)
for i in range(y_val.shape[1]):
    df_val[f'target_{i+1}'] = y_val[:, i]

df_test = pd.DataFrame(X_test)
for i in range(y_test.shape[1]):
    df_test[f'target_{i+1}'] = y_test[:, i]

print(df_train.describe())
print(df_val.describe())
print(df_test.describe())
```

```python
import matplotlib.pyplot as plt

df_train.hist(figsize=(20, 20))
plt.suptitle('Train Data Distributions')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(20, 10))
sns.boxplot(data=df_train)
plt.title('Box Plot of Train Data')
plt.show()
```

```python
plt.figure(figsize=(15, 12))
sns.heatmap(df_train.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Train Data')
plt.show()
```

```python
for i in range(y_train.shape[1]):
    plt.figure(figsize=(15, 6))
    plt.plot(df_train.index, df_train[f'target_{i+1}'], label=f'Train Target {i+1}')
    plt.plot(df_val.index, df_val[f'target_{i+1}'], label=f'Validation Target {i+1}')
    plt.plot(df_test.index, df_test[f'target_{i+1}'], label=f'Test Target {i+1}')
    plt.xlabel('Index')
    plt.ylabel(f'Target {i+1}')
    plt.title(f'Target {i+1} Over Time')
    plt.legend()
    plt.show()
```
